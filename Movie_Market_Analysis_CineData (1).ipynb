```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the movie dataset
df = pd.read_csv('movie_ratings_dataset.csv')
df.head()
```

```python
# Task 1: Calculate the average IMDB rating
avg_imdb = df['IMDB_Rating'].mean()
print(f"Overall Average IMDB Rating: {avg_imdb:.2f}")
```

```python
# Task 2: Compute average rating per genre and find the highest
genre_ratings = df.groupby('Genre')['IMDB_Rating'].mean().sort_values(ascending=False)
print("Average Rating per Genre:")
print(genre_ratings)

top_genre = genre_ratings.idxmax()
print(f"\nThe 'Genre King' with the highest rating is: {top_genre}")
```

```python
# Task 3: Bar Chart - Genre vs Box Office Revenue
plt.figure(figsize=(10, 6))
genre_revenue = df.groupby('Genre')['Box_Office_Millions'].sum().sort_values(ascending=False)

genre_revenue.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Box Office Revenue by Genre')
plt.xlabel('Genre')
plt.ylabel('Revenue (Millions $)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Task 4: Histogram of movie ratings
plt.figure(figsize=(10, 6))
plt.hist(df['IMDB_Rating'], bins=10, color='lightgreen', edgecolor='black')

plt.title('Distribution of Movie Ratings')
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency (Number of Movies)')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
# Task 5: Create 'Hit' column (True if Rating >= 8)
df['Hit'] = df['IMDB_Rating'] >= 8

# Task 6: Count total Hits and calculate percentage
hit_count = df['Hit'].sum()
hit_percentage = (hit_count / len(df)) * 100

print(f"Total Hit Movies: {hit_count}")
print(f"Hit Percentage: {hit_percentage:.1f}%")
df[['Genre', 'IMDB_Rating', 'Hit']].head()
```

```python
# Save the analyzed data to a new CSV file
df.to_csv('movie_analysis_final.csv', index=False)
print("Analysis complete and file saved!")
```
